# Hola aquí voy a declarar algunas funciones que despues vamos a nececitar:

```python
#funcion para convertir un numero a binario#
def binar(n):
    return(bin(n).replace("0b", ""))
#funcion para convertir un numero a binario#

#funcion para ver si un numero es capicúa#
def seracapi(n):
    if str(n) == str(n)[::-1]:
        return True
    else:
        return False
#funcion para ver si un numero es capicúa#
```

## La siguiente expresion nos genera una lista con los capicúas de nueve digitos

```python
capi9dig = [int("%d%d%d%d%d%d%d%d%d" % (a,b,c,d,e,d,c,b,a)) for a in range(1,10) for b in range(10) for c in range(10) for d in range(10) for e in range(10)]   
```

### Vamos a llamarla

```python
#1a)
capi9dig
```

### Oh que sorpresa! ahora miremos cuantos son, según la formula para capicúas de nueve dígitos el número debería ser $ 9 x10^{4} $ 

```python
#1a)
len(capi9dig)
```

### Ujuu, nos dió bien
### Ahora nos piden que miremos aquellos que son capicuas tambien en su representación binaria de la forma cap_base10 = cap_bin.

```python
#1b)
listbin10capi = []
for i in capi9dig:
    if seracapi(binar(i)) == True:
        listbin10capi.append(str(binar(i)) + " = " + str(i) )
listbin10capi
```

### Ahora vamos a ver que tan larga es la lista (aunque podemos ver fácilmente que la respuesta es 3)

```python
#1b)
len(listbin10capi)
```

### Ahora vamos a crear una funcion que revise si una ***cadena de caracteres*** es palíndroma o no:

```python
#1c)
def check_palin():
    n = raw_input("Introduce tu string: ")
    n_unicode=unicode(n,"UTF-8")
    listn = list(n_unicode)
    for i in range(len(listn)):
        if listn[i] == u'Á' or listn[i] == u'á':
            listn[i] = 'a'.encode('utf-8')
        elif listn[i] == u'É' or listn[i] == u'é':
            listn[i] = 'e'.encode('utf-8')
        elif listn[i] == u'Í' or listn[i] == u'í':
            listn[i] = 'i'.encode('utf-8')
        elif listn[i] == u'Ó' or listn[i] == u'ó':
            listn[i] = 'o'.encode('utf-8')
        elif listn[i] == u'Ú' or listn[i] == u'ú':
            listn[i] = 'u'.encode('utf-8')
        elif listn[i] == u'Ñ' or listn[i] == u'ñ':
            listn[i] = 'n'.encode('utf-8')    
        else:
            listn[i]=listn[i].encode('utf-8')  

    res = ''.join(map(str,listn))
    res = res.lower()
    
    res = res.replace("!","")
    res = res.replace(" ","")
    res = res.replace(",","")
    res = res.replace(".","")
    res = res.replace("?","")
    res = res.replace(":","")
    res = res.replace(";","")
    res = res.replace("","")
    res = res.replace("¡","")
    res = res.replace("@","")
    res = res.replace("#","")
    res = res.replace("$","")
    res = res.replace("%","")
    res = res.replace("^","")
    res = res.replace("&","")
    res = res.replace("*","")
    res = res.replace("(","")
    res = res.replace(")","")
    res = res.replace("-","")
    res = res.replace("=","")
    res = res.replace("+","")
    res = res.replace("/","")
    res = res.replace(">","")
    res = res.replace("<","")
    res = res.replace("'","")
    res = res.replace("{","")
    res = res.replace("}","")
    res = res.replace("|","")
    
    if str(res) == str(res)[::-1]:
        print "Es palindromo"
    else:
        return "No es palindromo"
    
```
